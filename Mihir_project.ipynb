```python
# Display the game board
def display_board(board):
    print("\n")
    print(f" {board[0]} | {board[1]} | {board[2]} ")
    print("---|---|---")
    print(f" {board[3]} | {board[4]} | {board[5]} ")
    print("---|---|---")
    print(f" {board[6]} | {board[7]} | {board[8]} ")
    print("\n")


# Check for win conditions
def check_win(board, player):
    winning_combinations = [
        [0, 1, 2],
        [3, 4, 5],
        [6, 7, 8],
        [0, 3, 6],
        [1, 4, 7],
        [2, 5, 8],
        [0, 4, 8],
        [2, 4, 6]
    ]

    for combo in winning_combinations:
        if board[combo[0]] == board[combo[1]] == board[combo[2]] == player:
            return True
    return False


# Check if board is draw
def is_draw(board):
    return " " not in board


# Main game loop
def play_game():
    board = [" "] * 9
    current_player = "X"

    print("üéÆ Welcome to Tic Tac Toe!")
    print("Player X goes first.")

    while True:
        display_board(board)

        # Player input
        try:
            move = int(input(f"Player {current_player}, choose a position (1-9): "))
        except ValueError:
            print("‚ùå Invalid input. Enter a number between 1 and 9.")
            continue

        if move < 1 or move > 9:
            print("‚ùå Position must be between 1 and 9.")
            continue

        if board[move - 1] != " ":
            print("‚ùå That spot is already taken!")
            continue

        # Apply move
        board[move - 1] = current_player

        # Check win
        if check_win(board, current_player):
            display_board(board)
            print(f"üèÜ Player {current_player} wins! üéâ")
            break

        # Check draw
        if is_draw(board):
            display_board(board)
            print("ü§ù It's a draw!")
            break

        # Switch player after one input
        current_player = "O" if current_player == "X" else "X"


# Start the game
play_game()
```
